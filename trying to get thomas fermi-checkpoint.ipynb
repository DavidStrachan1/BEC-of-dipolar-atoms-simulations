```python
import numpy as np
import scipy.constants as sc
from scipy.fft import fft,ifft,fftfreq
import matplotlib.pyplot as plt
from numpy import linalg as LA
import warnings
warnings.simplefilter('ignore')
from tqdm import tqdm_notebook
import mpl_toolkits.mplot3d as mplot3d
import scipy.special as ss
from scipy.integrate import quad, dblquad
# from mpl_tookits import mplot3d


N =300 
k = int(N/4)
alpha=np.pad(ss.jnp_zeros(0,N),(1,0),'constant',constant_values=(0)) # zeros of first derivative of zeroth order bessel function 
##starts from n=1 rather than n = 0

J0 = ss.jn_zeros(0,N+2) #zeros of zeroth order bessel function

S = ss.jv(0,alpha[k]*alpha/J0[N+1])**2
S /= ss.jv(0,alpha)**2
S = np.sqrt(1+sum(S)-S[0])
S /= np.abs(ss.jv(0,alpha[k]))/2



C = np.zeros((N+1,N+1))
for i in tqdm_notebook(range(N+1),leave = False):
    for j in range(N+1):
        C[i,j] = ss.jv(0,alpha[i]*alpha[j]/S)
        C[i,j] /= S*np.abs(ss.jv(0,alpha[i]))*np.abs(ss.jv(0,alpha[j]))/2

### write code for integral for cylindrical cut off         


# f = lambda x,y: x*np.cos(kz*y)*ss.jv(0,2*np.pi*rho*x)*(x**2 - 2*y**2)/(x**2 + y**2)**(5/2)
# # def U(rho,kz):
# #     out = dblquad(f,Rc,np.inf,lambda x: 0,lambda x: Zc)
# #     return out 
# # U = np.vectorize(U)


    

    
        
        
```

```python
from numpy.fft import fft2,ifft2
m = 2.7774e-25
w=100*np.pi

dt = -0.02j


b = 30#np.sqrt(S/(2*np.pi))
beta = S/(2*np.pi*b)#np.sqrt(S/(2*np.pi))
r = alpha*b/S
rho = alpha*beta/S

J = np.abs(ss.jv(0,alpha))

def F(f,J):
    out = f*b/J
    return out
def G(g,J):
    out = g*beta/J
    return out
def Finv(f,J):
    out = f*J/b
    return out
def Ginv(g,J):
    out = g*J/beta
    return out

######don't think this is doing what it should for 2d (including z)
def qfht(psi,J):
    out = F(psi,J) #changes variables
    out = np.matmul(C,out)#applies transformation matrix 
    out = Ginv(out,J)
    return out
def iqfht(psi,J):
    out = G(psi,J)
    out = np.matmul(C,out)
    out = Finv(out,J)
    return out

J = np.abs(ss.jv(0,alpha))


No = 1e5
##for harmonic
xs = np.sqrt(sc.hbar/(m*w))


a = 0#5.1e-9 
gs = 4*np.pi*a*No/xs #(not g, g in dimensionless units)


Rc = b+1
mu =6.98
mu_B =9.2740100783e-24
lamda = (m*No*sc.mu_0*(mu*mu_B)**2)/(3*(sc.hbar**2)*xs) 
# dimensionless variables


def dd_cont(psi):
    out = lamda/3
    out+= lamda*np.cos(Rc*2*np.pi*rho)/(Rc*np.pi*rho)**2
    out = np.nan_to_num(out)
    out = out-lamda*np.sin(Rc*2*np.pi*rho)/(Rc*2*np.pi*rho)**3
    out = np.nan_to_num(out)
    out*= -1 #ky/k_mag=cos(theta_k) where theta_k is the 
    #angle between k and the direction of the dipoles, I'm assuming y is 
    #the direction of the dipoles
    out = ifft(out*fft(psi**2))
    return out


#harmonic potential
def V(psi):
    return 0.5*r**2 + gs*psi**2 #+ dd_cont(psi)

psi = np.exp(-r**2)



# # # ##box potential
# box=r[int(N/5)] # in units of xs,slightly smaller than system
# def V(psi):
#     out = (1.01*r/box)**1000 #+ gs*np.abs(psi)**2 + dd_cont(psi)
#     return out


# def V(psi):
#     V_bools = np.abs(r)>=box
#     out = np.inf*V_bools
#     out = np.nan_to_num(out,posinf=np.inf)
#     out = out #+ gs*np.abs(psi)**2 + dd_cont(psi)
#     return out

    

# def psi_init(r):
#     out = ss.jv(0,J0[0]*r/box)+ss.jv(0,J0[1]*r/box) 
#     bools = np.abs(r)<box
#     return out*bools
# psi = psi_init(r)



T = 0.5*(2*np.pi*rho)**2
exp_T = np.exp(-1j*dt*T)

for i in tqdm_notebook(range(1000),leave = False):
    psi = np.exp(-0.5j*dt*V(psi))*psi
    psi = iqfht(exp_T*qfht(psi,J),J)
    psi = np.exp(-0.5j*dt*V(psi))*psi
    psi /= (LA.norm(psi))
psi *= No 


#harmonic potential    
gnd_state = np.exp((-r**2)/2)
gnd_state /= LA.norm(gnd_state)
gnd_state *= No

# ##box potential
# def gnd_state(r):
#     bessel_values = ss.jv(0,J0[0]*r/box)
#     bessel_bools = np.abs(r)<box
#     return bessel_values*bessel_bools
# gnd_state = gnd_state(r)
# gnd_state /= LA.norm(gnd_state)    
# gnd_state *= No


plt.plot(r,ifft(T),label="T")
plt.plot(r,0.5*r**2, label = "harmonic potential")
#plt.plot(r,gs*psi**2,label = "contact potential")
plt.legend()
#plt.ylim(0,5)
plt.title('potential')
plt.show()

plt.plot(r,np.abs(psi), label ="numerical ground state")
plt.plot(r,gnd_state, label = "true ground state")
plt.xlabel('r position')
plt.ylabel('psi')
#plt.xlim(0,20)
plt.legend()
plt.title('Psi against position')
plt.show() # plots psi against x
    
diff = psi-gnd_state
#plt.plot(r,100*diff/np.abs(psi), label ="difference to solution and true ground state")
#gives 100% at r[25] as gnd_state is 0 where as psi is finite (e-44)
plt.plot(r,100*diff/np.max(psi),label="diff")
plt.xlabel('r position')
plt.ylabel('diff/% of max of psi')
#plt.ylim(-1,1)
plt.xlim(0,10)
plt.legend()
plt.title('diff/%')
plt.show() # plots psi against x


##Thomas fermi 

# chem =sc.hbar*w*np.sqrt(4*No*a)
# chem /=(sc.hbar*w) #making it dimensionless 
# R = np.sqrt(2*chem) # thomas-fermi radius
# TF = np.sqrt(1-(r/R)**2)
# TF = np.nan_to_num(TF)
# TF /= LA.norm(TF)
# plt.plot(r,TF)

##R = 1e9 for R>1 (TF to be valid)

psi_r = psi
```
