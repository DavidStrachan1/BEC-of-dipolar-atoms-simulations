```python
import math
import numpy as np
import scipy.constants as sc
from scipy.fft import fft,ifft,fftfreq,rfftfreq
import matplotlib.pyplot as plt
from numpy import linalg as LA
import warnings
warnings.simplefilter('ignore')
from tqdm import tqdm_notebook
import seaborn as sns
import mpl_toolkits.mplot3d as mplot3d
from numpy.fft import fft2,ifft2,rfft2, irfft2, rfft,irfft

import scipy.special as ss
size = 30
N=4*size+1#has to be odd so there is a single max point
def E(psi,r,gs,eps,dx): ##see equation 1.42 in rigorous paper
    integrand = 0.5*np.abs(np.gradient(psi))**2 +0.5*(r*np.abs(psi))**2
    +0.5*gs*(np.abs(psi)**4)/np.sqrt(2*np.pi*eps) + 0.5*dd_quasi_2D_conv(psi,gamma,D,lamda,k_mag)*(np.abs(psi)**2) 
    integral = dx*dx*np.sum(integrand)
    return integral

def dd_quasi_2D_conv(psi,gamma,D,lamda,k_mag):
    lamda = 4*np.pi*D/3 ## keep this stuff outside the function so you don't constant calling it
    eps = 1/gamma
    xi = k_mag*np.sqrt(eps/2)
    out = 2-3*xi*np.sqrt(np.pi)*np.exp(xi**2)*ss.erfc(xi)
    out *= 1/(np.sqrt(2*np.pi*eps))
    #out = lamda*irfft(irfft(out*rfft(rfft(np.abs(psi)**2,axis=0),axis=1),N,axis=1),N,axis=0)
    out = lamda*ifft2(out*fft2(np.abs(psi)**2))
    
    return out 


def V(psi,gamma,gs):
    eps = 1/gamma
    return 0.5*(r**2) + (gs*np.abs(psi)**2)/np.sqrt(2*np.pi*eps)+dd_quasi_2D_conv(psi,gamma,D,lamda,k_mag)

    
```

```python
##should set N as a power of 2 to speed up fast fourier transforms
No = 1e5 #number of particles
m = 2.7774e-25
w = 100*np.pi
gamma = 2 #ratio of wz to w, large for large wz
dx = 2 * size/N
dt = -0.01j#-0.5j/gamma

x = np.linspace(-size,size,N)
x = x.reshape(-1,1).repeat(len(x),1)
y = x.T
#kx = rfftfreq(len(x),dx/(2*np.pi)).reshape(-1,1)
#kx=kx.repeat(len(kx),1) #defining the kx values
kx = fftfreq(len(x),dx/(2*np.pi)).reshape(-1,1).repeat(len(x),1)
ky = kx.T #defining the ky values
k_mag=np.sqrt(kx**2+ky**2)
T = 0.5*(kx**2+ky**2) #kinetic energy 2D array

##harmonic potential
xs = np.sqrt(sc.hbar/(m*w)) # scaling parameter of dimensionless length
##box potential
#xs = 1e-5
Ao = 5.29*1e-11 #Bohr radius
a = 100*Ao ##See Fortran paper, they analyse stability by varying a
gs =4*np.pi*a*No/xs 
g = 4*np.pi*(sc.hbar**2)*a/m
eps = 1/gamma ## matters for beta_2D

mu = 6.98 #dipole moment in units of Bohr magneton
mu_B =9.2740100783e-24
Cdd = sc.mu_0*(mu*mu_B)**2
lamda = (m*No*Cdd)/(3*(sc.hbar**2)*xs) 
# dimensionless variables
nz = 1 #dipoles have to be polarised along z for dd_quasi_conv term to be true, but Hankel is not
##used for this code, would be more efficient if it was but we need to know how to go into both fourier space
##in the z direction and bessel in the r direction simultaneously

R = (8*lamda/(np.pi*np.sqrt(2*np.pi*eps)))**0.25
R_validity = lamda/(2*np.pi*np.sqrt(gamma**3)) ##both from quasi 2D/1D paper, notation
##is confusing, compare to rigorous paper

D = 3*lamda/(4*np.pi)##book keeping for stability
#Edd = Cdd/(3*g)

r=np.sqrt(x**2+y**2) #gives the value of r for each point
##harmonic well

psi = np.exp(-(r**2)/10)
psi /= dx*LA.norm(psi)

exp_T = np.exp((-1j*dt*T))
for i in tqdm_notebook(range(4000),leave = False):
    psi = np.exp(-0.5j*dt*V(psi,gamma,gs))*psi
    #psi /= dx*LA.norm(psi)
    #psi = exp_T*rfft(rfft(psi,axis=0),axis=1)
    #psi = irfft(irfft(psi,N,axis=1),N,axis=0)
    psi = exp_T*fft2(psi)
    psi = ifft2(psi)
    psi /= dx*LA.norm(psi)
    psi = np.exp(-0.5j*dt*V(psi,gamma,gs))*psi
    psi /= dx*LA.norm(psi)
##it has to be in this order and separated like this to ensure psi 
#is in the right space for the functions
psi *= np.sqrt(No)

##harmonic ground state
gnd_state = np.exp((-(r**2)/2))
gnd_state /= dx*LA.norm(gnd_state)
gnd_state *= np.sqrt(No)
TF = 1-(r/R)**2
TF *= 2*No/(np.pi*(R**2))
TF = np.sqrt(TF) ##going from density to np.abs(psi)
diff = np.abs(psi)-np.abs(gnd_state)

midway_index = int((N-1)/2)
print(D)
print(gamma)
#print(Edd)
print(E(psi,r,gs,eps,dx))

#plt.plot(x[:,0],np.abs(psi[:,midway_index]), label ="numerical ground state, y=0")
plt.plot(x[:,0],psi[:,midway_index].real, label ="numerical ground state,y=0")
plt.plot(x[:,0],gnd_state[:,midway_index], label ="true ground state,y=0")
plt.plot(x[:,0],TF[:,midway_index], label ="Thomas Fermi, y=0")
#plt.xlim(-10,10)
plt.xlabel('position')
plt.ylabel('psi')    
plt.legend()
plt.title('Psi against position')
plt.show() 

plt.plot(x[:,0],100*diff[:,midway_index]/np.max(psi),label="diff, y=0")
plt.xlabel('r position')
plt.ylabel('diff/% of max of psi')
#plt.ylim(-1,1)
plt.xlim(0,10)
plt.legend()
plt.title('diff/%')
plt.show() # plots psi against x

fig = plt.figure()
ax = plt.axes(projection='3d')
bools = ((np.abs(x) <= 10)*(np.abs(y) <= 10)).ravel()##sets region for graph to show
n_ = int(np.sqrt(sum(bools)))
ax.plot_surface(x.ravel()[bools].reshape((n_,n_)),
                y.ravel()[bools].reshape((n_,n_)),
                np.abs(psi).ravel()[bools].reshape((n_,n_)),
                cmap='jet')
ax.set_title('ground state of 2D harmonic potential')
ax.set_xlabel('x position')
ax.set_ylabel('y position')
ax.set_zlabel('psi')
plt.show()

ax = plt.axes(projection='3d')
diff_perc = 100*diff/np.max(psi)
ax.plot_surface(x.ravel()[bools].reshape((n_,n_)),
                y.ravel()[bools].reshape((n_,n_)),
                np.abs(diff_perc).ravel()[bools].reshape((n_,n_)),
                cmap='jet')
ax.set_title('difference in percentage between gnd state and numerical solution')
ax.set_xlabel('x position')
ax.set_ylabel('y position')
ax.set_zlabel('psi')

fig,ax = plt.subplots()
ax.contour(x,y,np.abs(psi))
ax.set_ylim(-10,10)
ax.set_xlim(-10,10)
ax.set_xlabel('x position')
ax.set_ylabel('y position')
plt.show()

sns.heatmap(np.abs(psi))
plt.show()        
```

```python
gs = 0 #purely dipolar
Ds = np.linspace(0,5000,20)
gammas = np.linspace(1,50,20)
stability_matrix = np.zeros([len(gammas),len(Ds)])
init_psi = np.exp((-r**2)/10)
#init_psi /= Norm_r_or_rho(psiq_symm,r)    


p_max = 1000
for i in tqdm_notebook(range(len(gammas)),leave = False):
    for j in range(len(Ds)):
        hist_psi = [[0,init_psi[int(N/4):int(3*N/4),int(N/4):int(3*N/4)]]]
        gamma = gammas[i]
        D = Ds[j]
        isConv=False
        p=1
        
        #for p in tqdm_notebook(range(80),leave = False):
        while isConv==False and p<p_max:
            psi = np.exp(-0.5j*dt*V(psi,gamma,gs))*psi
            #psi /= dx*LA.norm(psi)
            psi = exp_T*fft2(psi)
            psi = ifft2(psi)
            psi /= dx*LA.norm(psi)
            psi *= np.exp(-0.5j*dt*V(psi,gamma,gs))
            psi /= dx*LA.norm(psi)
            
            hist_psi.append([p,psi[int(N/4):int(3*N/4),int(N/4):int(3*N/4)]]) ##potential efficiency problem, hist_psi increases in size 
            if len(hist_psi) > 100:
                hist_psi = hist_psi[-100:] ## gives last 2
                #mean_percent_diff = 100*np.mean((hist_psi[9][1]-hist_psi[0][1])/hist_psi[0][1])
                psi_max_percent_diff_array = np.zeros(10)
                for k in range(10):
                    psi_max_percent_diff_array[k] = 100*(np.max(hist_psi[10*k][1]) - np.max(hist_psi[0][1]))
                    psi_max_percent_diff_array[k] /= np.max(hist_psi[0][1])
                    
            
            
#             val_i1 = np.abs(np.mean(hist_psi[1][1])) ##
#             val_i0 = np.abs(np.mean(hist_psi[0][1]))
#             diff = val_i1-val_i0
#             percent_change = 100*np.abs(diff)/val_i0
                
                if np.ptp(psi_max_percent_diff_array)<0.001:
                    if np.where(hist_psi[0][1]==np.max(hist_psi[0][1]))[0][0] == np.where(hist_psi[-1][1]==np.max(hist_psi[-1][1]))[0][0]:
                        if (sorted(psi.ravel())[-1]+sorted(psi.ravel())[-2])/sorted(psi.ravel())[-1] > 1.1: ## if second largest value is greater than 10% of largest value
                            if (np.abs(psi)[midway_index,midway_index] == np.max(np.abs(psi))):
                                stability_matrix[i][j]=2 ##non red blood cell
                                isConv=True
                            else:
                                stability_matrix[i][j]=1 ##red blood cell
                                isConv=True
                if math.isnan(psi.any()) ==True:
                    isConv=True
            p+=1
        if stability_matrix[i][j]==0:
            psiq_symm = init_psi ##setting to original psi if psi hasn't converged

                
            
        
            
        #if p==p_max:
            #psi=init_psi
            

            
            
gammav = gammas.reshape(-1,1).repeat(len(Ds),1)
Dsv = Ds.reshape(1,-1).repeat(len(gammas),0)

sns.heatmap(stability_matrix)
plt.show() 

fig=plt.figure()
ax=plt.axes(projection="3d")
ax.plot_surface(gammav,Dsv,stability_matrix,cmap="jet")

plt.xlabel("gamma")
plt.ylabel("Dipole strength")
ax.set_title("Convergence plot")
plt.show()                           
```
