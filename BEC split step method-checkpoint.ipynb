# 1D split step method with harmonic external potential and no contact potential

```python
import numpy as np
import scipy.constants as sc
from scipy.fft import fft,ifft,fftfreq
import matplotlib.pyplot as plt
from numpy import linalg as LA
import warnings
warnings.simplefilter('ignore')
from tqdm import tqdm_notebook
N = 3999 # number of x values, increasing this increases the accuracy of the solution

m = 2.7774e-25 # mass of erbium in kg
w = 20*np.pi #angular frequency

dt = -0.0001j #imaginary time propagation in dimensionless units
dx = 1e-3 #sample spacing
x = np.linspace(-N*dx/2,N*dx/2,N) #setting the x values in dimensionless units
k = fftfreq(len(x),dx/(2*np.pi)) # setting the k values

T = 0.5*(k**2) #kinetic energy array

No = 3e5 #number of particles

##harmonic potential
xs = np.sqrt(sc.hbar/(m*w)) # scaling parameter of dimensionless length 

##box potential
#xs = 1e-5 #size of box
a= 5.1e-9 ##s wave scattering length
gs = 4*np.pi*a*No/xs 

Rc = 3 #spherical cut-off, greater than the system size so 
#is physically resonable
mu = 6.98 #dipole moment in units of Bohr magneton
mu_B =9.2740100783e-24
lamda = (m*No*sc.mu_0*(mu*mu_B)**2)/(3*(sc.hbar**2)*xs) 
# dimensionless variables

### dipole-dipole interaction potential  

def dd_cont(psi):
    out = lamda/3
    out+= lamda*np.cos(Rc*k)/(Rc*k)**2-lamda*np.sin(Rc*k)/(Rc*k)**3
    out = 2*np.nan_to_num(out) #factor of two comes from theta_k being 0 
    #as this is 1D
    out = ifft(out*fft(psi**2))
    return out



##harmonic potential
# def V(psi):
#     out = 0.5*x**2 #+gs*np.abs(psi)**2 +dd_cont(psi)
#     return out
# psi =np.exp(-x**2) #initial guess of gaussian


##box potential

def V(psi):
    out = x**1000 + gs*psi**2 + dd_cont(psi)
    return out

def psi_init(x):
        cos_values = 1#np.cos(np.pi*x/2)+2*np.cos(3*np.pi*x/2)
        cos_bools = (np.abs(x)<1)
        return cos_values*cos_bools
psi = psi_init(x)
    



exp_T =np.exp(-1j*dt*T) # combining exp(T) and (imaginary) time propagation


for i in tqdm_notebook(range(1000),leave = False):  
    psi = np.exp(-0.5j*dt*V(psi))*ifft(exp_T*fft(np.exp(-0.5j*dt*V(psi))*psi))

psi /= LA.norm(psi) #normalisation

##harmonic ground state
gnd_state = np.exp((-x**2)/2)

##box potential ground state
def gnd_state(x):
        cos_values = np.cos(np.pi*x/2)
        cos_bools = (np.abs(x)<1)
        return cos_values*cos_bools
gnd_state = gnd_state(x)
    





gnd_state /= LA.norm(gnd_state)
diff = psi-gnd_state
print(np.max(np.abs(diff)))

plt.plot(x,psi, label ="numerical ground state")
plt.plot(x,gnd_state, label ="true ground state")
plt.xlim(-2,2)
plt.xlabel('position')
plt.ylabel('psi')    
plt.legend()
plt.title('Psi against position')
plt.show() # plots psi against x
```

# 2D split step method with harmonic external potential and contact potential

```python
import seaborn as sns
from mpl_toolkits import mplot3d
from numpy.fft import fft2,ifft2

N=499

m = 2.7774e-25
w = 20*np.pi

dx = 1e-1
dt = -0.001j
x = np.linspace(-N*dx/2,N*dx/2,N)
x = x.reshape(-1,1).repeat(len(x),1)
y = x.T


kx = fftfreq(len(x),dx/(2*np.pi)).reshape(-1,1).repeat(len(x),1) #defining the kx values
ky = kx.T #defining the ky values
k_mag=np.sqrt(kx**2+ky**2)

T = 0.5*(kx**2+ky**2) #kinetic energy 2D array
exp_T =np.exp((-1j*dt*T)) # combining exp(T) and (imaginary) time propagation


No = 3e5
##harmonic potential
xs = np.sqrt(sc.hbar/(m*w)) # scaling parameter of dimensionless length

##box potential
#xs = 1e-5
a = 5.1e-9
gs =0#4*np.pi*a*No/xs 





Rc = 25 #spherical cut-off, greater than the system size so 
#is physically resonable
mu = 6.98 #dipole moment in units of Bohr magneton
mu_B =9.2740100783e-24
lamda = (m*No*sc.mu_0*(mu*mu_B)**2)/(3*(sc.hbar**2)*xs) 
# dimensionless variables




def dd_cont(psi):
    out = lamda/3
    out+= lamda*np.cos(Rc*k_mag)/(Rc*k_mag)**2-lamda*np.sin(Rc*k_mag)/(Rc*k_mag)**3
    out = np.nan_to_num(out)
    out*= 3*(np.nan_to_num((ky/k_mag)**2))-1 #ky/k_mag=cos(theta_k) where theta_k is the 
    #angle between k and the direction of the dipoles, I'm assuming y is 
    #the direction of the dipoles
    out = ifft2(out*fft2(psi**2))
    return out


##harmonic well
def V(psi):
    return 0.5*(x**2+y**2) + gs*np.abs(psi)**2 +dd_cont(psi)

psi =np.exp(-(x**2+y**2)) #initial guess for harmonic potential

# #circular box potential
# def V(psi):
#     return  (x**2+y**2)**100 + gs*psi**2 +dd_cont(psi)

    
# def psi_init(x): # initial guess for circular box potential
#     out = 1
#     cos_bools = (np.abs(x**2+y**2)<1)
#     return out*cos_bools
# psi = psi_init(x)
    



for i in tqdm_notebook(range(100),leave = False):
    
    
    psi = np.exp(-0.5j*dt*V(psi))*ifft2(exp_T*fft2(np.exp(-0.5j*dt*V(psi))*psi))
    psi /= LA.norm(psi)

##harmonic ground state
gnd_state = np.exp((-(x**2+y**2)/2))

##circular box potential ground state
#agofpkngofngpo


gnd_state /= LA.norm(gnd_state)
diff = np.abs(psi)-np.abs(gnd_state)
print(np.mean(np.abs(diff)))

r_array = np.sqrt(x**2+y**2) #gives the value of r for each x,y coordinate
midway_index = int((N-1)/2)

plt.plot(r_array[midway_index,midway_index:],psi[midway_index,midway_index:], label ="numerical ground state")
plt.plot(r_array[midway_index,midway_index:],gnd_state[midway_index,midway_index:], label ="true ground state")
plt.xlim(0,10)
plt.xlabel('position')
plt.ylabel('psi')    
plt.legend()
plt.title('Psi against position')
plt.show() # plots psi against x


fig = plt.figure()
ax = plt.axes(projection='3d')

bools = ((np.abs(x) <= 5)*(np.abs(y) <= 5)).ravel()
n_ = int(np.sqrt(sum(bools)))
ax.plot_surface(x.ravel()[bools].reshape((n_,n_)),
                y.ravel()[bools].reshape((n_,n_)),
                np.abs(psi).ravel()[bools].reshape((n_,n_)),
                cmap='jet')
ax.set_title('ground state of 2D harmonic potential')
# ax.set_ylim(-5,5)
# ax.set_xlim(-5,5)
# plt.xlim(-5,5)
# plt.ylim(-5,5)
ax.set_xlabel('x position')
ax.set_ylabel('y position')
ax.set_zlabel('psi')

plt.show()


fig,ax = plt.subplots()
ax.contour(x,y,np.abs(psi))
ax.set_ylim(-5,5)
ax.set_xlim(-5,5)
ax.set_xlabel('x position')
ax.set_ylabel('y position')
plt.show()
# sns.heatmap(np.abs(diff))

# plt.show()
```
