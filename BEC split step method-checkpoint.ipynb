# 1D split step method with harmonic external potential and no contact potential

```python
import numpy as np
import scipy.constants as sc
from scipy.fft import fft,ifft,fftfreq
import matplotlib.pyplot as plt
from numpy import linalg as LA
import warnings
warnings.simplefilter('ignore')
from tqdm import tqdm_notebook
N = 1999 # number of x values

m = 2.7774e-25 # mass of erbium in kg
w = 20*np.pi #angular frequency

dt = -0.05j #imaginary time propagation in dimensionless units
dx = 1e-2 #sample spacing
x = np.linspace(-N*dx/2,N*dx/2,N) #setting the x values in dimensionless units
k = fftfreq(len(x),dx/(2*np.pi)) # setting the k values

T = 0.5*(k**2) #kinetic energy array

n = 3e5
xs = np.sqrt(sc.hbar/(m*w)) # scaling parameter of dimensionless length
a= 5.1e-9
gs = 4*np.pi*a*n/xs 

##harmonic potential
def V(psi):
    out = 0.5*x**2 +gs*np.abs(psi)**2
    return out

##box potential

#def V(psi):
   # out = x**100 + gs*psi**2
   # return out

psi =np.exp(-x**2) #initial guess of gaussian



exp_T =np.exp(-1j*dt*T) # combining exp(T) and (imaginary) time propagation

for i in tqdm_notebook(range(1000),leave = False):

    #psi = exp_V*ifft(exp_T*fft(exp_V*psi)) # loop that applies the split step time method
    psi = np.exp(-0.5j*dt*V(psi))*ifft(exp_T*fft(np.exp(-0.5j*dt*V(psi))*psi))
    psi /= LA.norm(psi) #normalisation


gnd_state = np.exp((-x**2)/2)
gnd_state /= LA.norm(gnd_state)
diff = psi-gnd_state
print(np.max(diff))

plt.plot(x,psi, label ="numerical ground state")
plt.plot(x,gnd_state, label ="true ground state")

plt.xlabel('position')
plt.ylabel('psi')    
plt.legend()
plt.title('Psi against position')
plt.show() # plots psi against x
```

# 3D split step method for harmonic external potential with contact potential



```python
from scipy.fft import fftn,ifftn

N = 100
dx=5e-2

x,y,z = np.linspace(-N*dx/2,N*dx/2,N),np.linspace(-N*dx/2,N*dx/2,N),np.linspace(-N*dx/2,N*dx/2,N)
kx,ky,kz = fftfreq(len(x),dx/(2*np.pi)),fftfreq(len(y),dx/(2*np.pi)),fftfreq(len(z),dx/(2*np.pi))

x,y,z = x.reshape((-1,1,1)),y.reshape((1,-1,1)),z.reshape((1,1,-1)) #turns them into 3D arrays, N by 1 by 1
#-1 is shorthand for N-1, means the size of the x direction will be N (0 to N-1)
kx,ky,kz = kx.reshape((-1,1,1)),ky.reshape((1,-1,1)),kz.reshape((1,1,-1))

x,y,z = x.repeat(N,1).repeat(N,2),y.repeat(N,0).repeat(N,2),z.repeat(N,0).repeat(N,1) #repeats the column along the other two 
#axes, so they all become N by N by N arrays with each varying in only one direction
kx,ky,kz =kx.repeat(N,1).repeat(N,2),ky.repeat(N,0).repeat(N,2),kz.repeat(N,0).repeat(N,1)


g=10

psi = np.exp(-(x**2+y**2+z**2)) #initial guess of gaussian


Rc = 10 #spherical cut-off, greater than the system size so is physically resonable
Cdd = 10
k_mag=np.sqrt(kx**2+ky**2+kz**2)
def dd_cont(psi):
    out = Cdd/3
    out+= Cdd*np.cos(Rc*k_mag)/(Rc*k_mag)**2-Cdd*np.sin(Rc*k_mag)/(Rc*k_mag)**3
    out = np.nan_to_num(out)
    out*= 3*(np.nan_to_num((kx/k_mag)**2))-1 #kx/k_mag=cos(theta_k) where theta_k is the 
    #angle between k and the direction of the dipoles, I'm assuming x is 
    #the direction of the dipoles
    out = ifftn(out*fftn(psi**2))
    return out




def V(psi):
    return 0.5*m*w*(x**2+y**2+z**2) + g*psi**2 +dd_cont(psi)


#exp_V = np.exp(-1j*dt*V/(2*sc.hbar))

exp_T = np.exp((-1j*sc.hbar*dt*(kx**2+ky**2+kz**2)/2*m))

n=3

for i in tqdm_notebook(range(100),leave = False):
    psi = np.exp(-1j*dt*V(psi)/2)*ifftn(exp_T*fftn(np.exp(-1j*dt*V(psi)/2)*psi))

psi /= LA.norm(psi) 

fig = plt.figure()
ax = plt.axes(projection='3d')
q = int(N/2+float(input("Zvalue = "))/dx) #converts z value to the corresponding element in the x,y,psi arrays


ax.plot_surface(x[:,:,q], y[:,:,q], psi[:,:,q].real, cmap='jet')
ax.set_title('ground state of 3D harmonic potential')

ax.set_xlabel('x position (m)')
ax.set_ylabel('y position (m)')
ax.set_zlabel('psi')

plt.show()

sns.heatmap((psi.real).sum(axis=2))
    

```
