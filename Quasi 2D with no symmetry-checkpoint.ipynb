```python
import numpy as np
import scipy.constants as sc
from scipy.fft import fft,ifft,fftfreq
import matplotlib.pyplot as plt
from numpy import linalg as LA
import warnings
warnings.simplefilter('ignore')
from tqdm import tqdm_notebook
import seaborn as sns
#import mpl_toolkits as mplot3d
import mpl_toolkits.mplot3d as mplot3d
from numpy.fft import fft2,ifft2

N=500
m = 2.7774e-25
w = 100*np.pi
size = 10 #max radius
dx = 2 * size/N
dt = -0.1j
x = np.linspace(-size,size,N)
x = x.reshape(-1,1).repeat(len(x),1)
y = x.T
kx = fftfreq(len(x),dx/(2*np.pi)).reshape(-1,1).repeat(len(x),1) #defining the kx values
ky = kx.T #defining the ky values
k_mag=np.sqrt(kx**2+ky**2)
T = 0.5*(kx**2+ky**2) #kinetic energy 2D array
exp_T =np.exp((-1j*dt*T)) # combining exp(T) and (imaginary) time propagation


No = 3e5
##harmonic potential
xs = np.sqrt(sc.hbar/(m*w)) # scaling parameter of dimensionless length
##box potential
#xs = 1e-5
a = 5.1e-9
gs =4*np.pi*a*No/xs 


Rc = 25 #spherical cut-off, greater than the system size so 
#is physically resonable
mu = 6.98 #dipole moment in units of Bohr magneton
mu_B =9.2740100783e-24
lamda = (m*No*sc.mu_0*(mu*mu_B)**2)/(3*(sc.hbar**2)*xs) 
# dimensionless variables
nx = 0
ny = 0 
nz = 1 #components of normalised vector of magnetic polarisation
gamma = 10 #ratio of wz to w, large for large wz
eps = 1/gamma
beta_2D = gs-lamda+3*lamda*(nz**2)
beta_2D /= eps*(2*np.pi)**0.5


# def dd_cont(psi):
#     out = lamda/3
#     out+= lamda*np.cos(Rc*k_mag)/(Rc*k_mag)**2-lamda*np.sin(Rc*k_mag)/(Rc*k_mag)**3
#     out = np.nan_to_num(out)
#     out*= -1 #ky/k_mag=cos(theta_k) where theta_k is the 
#     #angle between k and the direction of the dipoles, I'm assuming z is 
#     #the direction of the dipoles
#     out = ifft2(out*fft2(psi**2))
#     return out

r=np.sqrt(x**2+y**2) #gives the value of r for each x,y coordinate
##harmonic well
def V(psi):
    return 0.5*(r**2) + beta_2D*np.abs(psi)**2 #+dd_cont(psi)

psi =np.exp(-(r**2)) #initial guess for harmonic potential

# #circular box potential

  
# def V(psi):
#     V_bools = np.abs(r)>=box
#     out = np.inf*V_bools
#     out = np.nan_to_num(out,posinf=np.inf)
#     out = out + gs*np.abs(psi)**2 + dd_cont(psi)
#     return out

# def psi_init(x):
#         cos_values = 1#np.cos(np.pi*x/2*box)+2*np.cos(3*np.pi*x/2*box)
#         cos_bools = (np.abs(r)<box)
#         return cos_values*cos_bools
# psi = psi_init(x)

def dd_quasi_2D(psi):
    out = ((nz**3 -nx**2)*kx**2 + (nz**3 -ny**2)*ky**2)*(psi**2)/k_mag
    out = np.nan_to_num(out)
    out *= 3*lamda/2 #has to be positive so exponential decays,
    #but I must be a factor of -1 out somewhere 
    
    return out

def exp_fourier_quasi_2D(psi): # includes kinetic and dd contribution
    return np.exp(-1j*dt*(dd_quasi_2D(psi)+T))

# for i in tqdm_notebook(range(100),leave = False):
    
    
#     psi = np.exp(-0.5j*dt*V(psi))*ifft2(exp_T*fft2(np.exp(-0.5j*dt*V(psi))*psi))
#     psi /= LA.norm(psi)
# psi *= No


for i in tqdm_notebook(range(59),leave = False):
    
    
    psi = fft2(np.exp(-0.5j*dt*V(psi))*psi)
    psi *= exp_fourier_quasi_2D(psi)
    psi = np.exp(-0.5j*dt*V(psi))*ifft2(psi)
    psi /= LA.norm(psi)
psi *= No

##harmonic ground state
gnd_state = np.exp((-(r**2)/2))



gnd_state /= LA.norm(gnd_state)
gnd_state *= No
diff = np.abs(psi)-np.abs(gnd_state)

midway_index = int((N-1)/2)

plt.plot(r[midway_index,midway_index:],psi[midway_index,midway_index:], label ="numerical ground state")
plt.plot(r[midway_index,midway_index:],gnd_state[midway_index,midway_index:], label ="true ground state")
plt.xlim(0,10)
plt.xlabel('position')
plt.ylabel('psi')    
plt.legend()
plt.title('Psi against position')
plt.show() # plots psi against x


fig = plt.figure()
ax = plt.axes(projection='3d')

bools = ((np.abs(x) <= 5)*(np.abs(y) <= 5)).ravel()
n_ = int(np.sqrt(sum(bools)))
ax.plot_surface(x.ravel()[bools].reshape((n_,n_)),
                y.ravel()[bools].reshape((n_,n_)),
                np.abs(psi).ravel()[bools].reshape((n_,n_)),
                cmap='jet')
ax.set_title('ground state of 2D harmonic potential')
# ax.set_ylim(-5,5)
# ax.set_xlim(-5,5)
# plt.xlim(-5,5)
# plt.ylim(-5,5)
ax.set_xlabel('x position')
ax.set_ylabel('y position')
ax.set_zlabel('psi')

plt.show()

ax = plt.axes(projection='3d')
diff_perc = 100*diff/np.max(psi)
ax.plot_surface(x.ravel()[bools].reshape((n_,n_)),
                y.ravel()[bools].reshape((n_,n_)),
                np.abs(diff_perc).ravel()[bools].reshape((n_,n_)),
                cmap='jet')
ax.set_title('difference in percentage between gnd state and numerical solution')

ax.set_xlabel('x position')
ax.set_ylabel('y position')
ax.set_zlabel('psi')


fig,ax = plt.subplots()
ax.contour(x,y,np.abs(psi))
ax.set_ylim(-5,5)
ax.set_xlim(-5,5)
ax.set_xlabel('x position')
ax.set_ylabel('y position')
plt.show()
```
